```python
import speech_recognition as sr
from googletrans import Translator
from gtts import gTTS
from playsound import playsound
import os
```

```python
while True:
    
    r = sr.Recognizer()
    translator =  Translator()
        
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source)
        
        print("please say some thing... ")
        audio = r.listen(source)
        
        
        try:
            text = r.recognize_google(audio)
               
            print("you have said:    "  + text)
            if(text == "exit"):
                break
            translated_text = translator.translate(text, dest='hi')
           
            print(translated_text.text)
            voice = gTTS(translated_text.text,lang='hi')
            voice.save('voice.mp3')
            playsound('voice.mp3')
            os.remove('voice.mp3')

            
        except sr.UnknownValueError:
                  
            print("could not understand : \n Please Repeat again")
            
        except sr.RequestError:
                  
            print("could not Request result from google")

 
```
